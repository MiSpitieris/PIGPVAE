```python
import torch
import numpy as np
from torchdiffeq import odeint
import matplotlib.pyplot as plt
import torch.nn as nn
```

```python
seed = 42
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# Define model classes (unchanged from your definition)
class DampedDrivenPendulum(torch.nn.Module):
    def __init__(self, omega):
        super().__init__()
        self.omega = omega

    def forward(self, t, y):
        theta, theta_dot = y
        # gamma = 0.15
        gamma = 0.15
        F0 = 0.5
        omega_drive = 2.0
        return torch.stack([
            theta_dot,
            -self.omega * torch.sin(theta) - gamma * theta_dot + F0 * torch.cos(omega_drive * t)
        ])

class DampedPendulum(torch.nn.Module):
    def __init__(self, omega, gamma=0.15):
        super().__init__()
        self.omega = omega
        self.gamma = gamma

    def forward(self, t, y):
        theta, theta_dot = y
        return torch.stack([
            theta_dot,
            -self.omega * torch.sin(theta) - self.gamma * theta_dot
        ])

class SimplePendulum(nn.Module):
    def __init__(self):
        super(SimplePendulum, self).__init__()

    def forward(self, t, y, omega):
        """
        Parameters:
            t (torch.Tensor): The time tensor.
            y (tuple of torch.Tensor): Tuple containing (theta, theta_dot).

        Returns:
            torch.Tensor: Time derivatives of theta and theta_dot.
        """
        theta, theta_dot = y
        return torch.stack([
            theta_dot,
            -omega * torch.sin(theta)
        ])
# Define time span
t_span = torch.linspace(0, 10, 100)

# Define omega values
omega_values = np.linspace(0.8, 1.2, 20)
# omega_values = np.linspace(1.5, 2.5, 20)
simulated_data_simple = {}
simulated_data_damped = {}

noise_std = 0.05

for omega in omega_values:
    # Random initial conditions for each omega
    theta_initial = np.random.uniform(-1, 1)
    theta_dot_initial = 0
    theta_0 = torch.tensor([theta_initial, theta_dot_initial])

    # simple_model = SimplePendulum(torch.tensor(omega))
    simple_model = SimplePendulum()
    damped_model = DampedDrivenPendulum(torch.tensor(omega))

    key = (omega, theta_0[0].item(), theta_0[1].item())

    with torch.no_grad():
        sol_simple = odeint(func=lambda t, y: simple_model(t, y, omega), y0=theta_0, t=t_span)[:, 0]
        sol_damped = odeint(damped_model, theta_0, t_span)[:, 0]
        noise = torch.randn_like(sol_damped) * noise_std
        sol_damped_noisy = sol_damped + noise

    simulated_data_simple[key] = sol_simple
    # simulated_data_damped[key] = sol_damped
    simulated_data_damped[key] = sol_damped_noisy
```

```python
# Calculate number of rows and columns for subplots
num_cols = 2
num_rows = len(omega_values) // num_cols + (len(omega_values) % num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(14, num_rows * 3.5))
axes = axes.flatten()

for i, omega in enumerate(omega_values):
    ax = axes[i]
    
    # Retrieve the corresponding initial conditions
    key = [k for k in simulated_data_simple if k[0] == omega][0]
    sol_simple = simulated_data_simple[key]
    sol_damped = simulated_data_damped[key]

    theta_init, theta_dot_init = key[1], key[2]

    ax = axes[i]
    ax.plot(t_span.numpy(), sol_simple.numpy(), '--', label='Simple Pendulum')
    ax.plot(t_span.numpy(), sol_damped.numpy(), '-', label='Damped Driven Pendulum')
    ax.set_title(f'ω = {omega:.2f}, θ₀={theta_init:.2f}, θ̇₀={theta_dot_init:.2f}')
    ax.set_xlabel('Time')
    ax.set_ylabel(r'$\theta$')
    ax.legend(fontsize='small')
    ax.grid(True)

# Remove any unused subplots
for j in range(i + 1, num_rows * num_cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for key, sol_damped in simulated_data_damped.items():
    omega, theta_init, theta_dot_init = key
    plt.plot(t_span.numpy(), sol_damped.numpy(),  alpha=0.7)

plt.xlabel('time')
plt.ylabel('Angle (θ)')
plt.title('Damped Driven Pendulum Trajectories\n for different ω and initial conditions')
plt.grid(True)
plt.tight_layout()
plt.show()
```
