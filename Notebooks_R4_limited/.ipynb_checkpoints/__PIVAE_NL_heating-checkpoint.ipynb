# VAE with Newton's law decoder

The following code defines a VAE where the decoder model is a physical model simulator and not a NN. The idea is to use Netwon's law of heating as an approximation of the underlying physics to generate plausible room temperature data during heating (equivallently we can do the same for cooling).

The Newton's law of cooling (heating for our application)  states that the temperature of a body changes at a rate proportional to the difference in temperature between the body and its surroundings (source http://amsi.org.au/ESA_Senior_Years/SeniorTopic3/3e/3e_4history_3.html).

It is mathematically described by the following differential equation:
$$
\dfrac{dT}{dt} = -k (T - T_s) \qquad  (1)
$$
where $T$ is the temperature, $T_s$ is the temperature of the surroundings (e.g. outer temperature) and $k$ is a constant that controls the rate of temperature increase (heating) or decrease (cooling) in time. Therefore, the $k$ parameter is has physical interpretation.

Equation 1 is first-order and has the following trivial solution
$$
(T_0 - T_s)  \exp(-k\cdot t) + T_s \qquad  (2). 
$$
Therefore, it is fast to simulate data from the model.


Equation 2 will be the model decoder and we will infer the distribution of $k$ parameter from the data. Therefore, the latent parameter has direct physical meaning compared to standard VAEs where the decoder is a NN.

```python
import os
import sys

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import gpytorch
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import copy
from torch.utils.data import Dataset, DataLoader,random_split, Subset
from tqdm import tqdm
import random

sys.path.append(os.path.abspath('..'))
from Models.PIVAE import PIVAE
from physics import NewtonsLaw
```

```python
# Import RICO data 
file_path = '../Data/RICO4_Dataset_processed.hdf'
df = pd.read_hdf(file_path)
#  define surrounding temperature
df.loc[:, 'sur_temp'] = (df['RTD420'] + df['RTD417'] + df['B.ASTRHT2.T'])/3 
df_h = df.copy()
df_h = df[df['temp_change_class'] == 'h']
def select_points(group):
    return group.iloc[::10]  # Select every 10th row
df_h = df_h.groupby('Scheduler Step').apply(select_points).reset_index(drop=True)
df_h = df_h.dropna()
```

```python
class RICO4Dataset(Dataset):
    def __init__(self, df, select_var, exclude_samples=None):
        """
        Args:
            df: DataFrame containing the data.
            select_var: List of variables to include in the dataset.
            exclude_samples: List of samples (intervals) to exclude from the dataset.
        """
        self.df = df
        self.select_var = select_var
        if exclude_samples:
            self.unique_batches = [batch for batch in df['interval'].unique() if batch not in exclude_samples]
        else:
            self.unique_batches = df['interval'].unique()

    def __len__(self):
        return len(self.unique_batches)

    def __getitem__(self, idx):
        batch = self.unique_batches[idx]
        temp = self.df[self.df['interval'] == batch][self.select_var]
        
        # Normalize and convert to tensors
        ti = torch.tensor(temp['time_within_interval'].to_numpy() / 240, dtype=torch.float32)
        Ts = torch.tensor(temp['sur_temp'].to_numpy(), dtype=torch.float32)
        heat_obs = torch.tensor(temp['B.RTD1'].to_numpy(), dtype=torch.float32)
        
        # Stack the tensors to create the input batch
        x_batch = torch.stack((heat_obs, Ts, ti), dim=1)
        
        return x_batch

# Initialize the dataset, excluding samples 53 and 5
exclude_samples = [53, 5]
select_var = ['time_within_interval', 'sur_temp', 'B.RTD1']
dataset = RICO4Dataset(df_h, select_var, exclude_samples=exclude_samples)
dataloader = DataLoader(dataset)
# Validate the updated dataset
print(f"Dataset length after exclusion: {len(dataset)}")


# Define split ratios
train_ratio = 0.7
test_ratio = 1-train_ratio

# Calculate split sizes
total_size = len(dataset)
train_size = int(train_ratio * total_size)
test_size = total_size - train_size

# Define a fixed seed for reproducibility

seed = 0
generator = torch.Generator().manual_seed(seed)

# Split the dataset
train_subset, test_subset = random_split(
    dataset,
    [train_size, test_size],
    generator=generator
)
train_loader = DataLoader(train_subset, batch_size=1, shuffle=False)
val_loader = DataLoader(test_subset, batch_size=1, shuffle=False)
```

```python
# Set the variables to be selected from the dataframe
torch.manual_seed(0)
model = PIVAE(hidden_layers=[10,10], 
                 activation=nn.Tanh(),
                 mu_prior=torch.tensor(0),
                 var_prior=torch.tensor(2.0))
# Initialize model parameters and optimizer
pars = model.parameters()
opt = torch.optim.Adam(pars, lr=0.0001)

# Lists to store loss values
l_loss, KL, mse = [], [], []

# Set beta value and number of epochs
beta_phy = 1
num_epochs = 200
# epochs = range(3000)
l_loss, KL, mse = model.fit(dataloader, opt, beta_phy, num_epochs)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
# Loss plot
epochs = range(num_epochs)
axs[0].plot(epochs, l_loss, 'r-')
axs[0].set_title('Loss')
axs[0].set_xlabel('Epoch')
# axs[0].set_ylabel('Loss')

# MSE plot
axs[1].plot(epochs, mse, 'g-')
axs[1].set_title('MSE')
axs[1].set_xlabel('Epoch')
# axs[1].set_ylabel('MSE')

# KL Divergence plot
axs[2].plot(epochs, KL, 'b-')
axs[2].set_title('KL')
axs[2].set_xlabel('Epoch')
# axs[2].set_ylabel('KL')

plt.tight_layout()
plt.show()
```

```python
x_hat, mu, var = {}, {}, {} 
for batch in df_h['interval'].unique():  # Iterate over each mini-batch
        id = df_h['interval'] == batch
        temp = df_h[id][select_var]
        ti = torch.tensor(temp['time_within_interval'].to_numpy()/240, dtype=torch.float32)
        # ti = torch.tensor(temp['time_within_interval'].to_numpy(), dtype=torch.float32)
        Ts = torch.tensor(temp['sur_temp'].to_numpy(), dtype=torch.float32)
        heat_obs = torch.tensor(temp['B.RTD1'].to_numpy(), dtype=torch.float32)
        x_batch = torch.stack((heat_obs, Ts, ti), dim=1)
        with torch.no_grad():
            x_hat[batch], mu[batch], var[batch] = model(x_batch)
```

```python
# Convert mu dictionary to a tensor
mu_tensor = torch.cat([mu[batch] for batch in mu.keys()], dim=0)

# Convert var dictionary to a tensor
var_tensor = torch.cat([var[batch] for batch in var.keys()], dim=0)
```

```python
def generate_new_data(vae, num_samples, T0, t, Ts):
    vae.eval()  
    
    # Sample from a standard normal distribution
    z_samples = torch.randn(num_samples, 1)
    
    # Reparameterize (without noise here as we are generating new data)
    k_samples = torch.exp(z_samples)  # Apply the exponential to ensure positivity if required
    
    # Decode the samples to generate new data
    generated_data = []
    for k in k_samples:
        x_hat = vae.decode(T0, t, Ts, k.unsqueeze(0))
        generated_data.append(x_hat)
    
    return torch.stack(generated_data)

# Example usage
T0 = torch.tensor(10.0)  # Initial temperature or other relevant start value
t = ti # Time or other domain-specific input
Ts = torch.tensor(20.0)  # Mean of some distribution or parameter, specific to your model setup
new_data = generate_new_data(vae = PIVAE(), num_samples=1000, T0=T0, t=t, Ts=Ts)
```

```python
def generate_new_data(NL, num_samples, mu_post, var_post, T0, t, Ts):
    # vae.eval()  
    # Sample from a standard normal distribution
    epsilon = torch.randn(num_samples, 1)
    z_samples = mu_post + var_post.sqrt()*epsilon
    k_samples = z_samples.exp()
    
    # Decode the samples to generate new data
    generated_data = []
    for k in k_samples:
        x_hat = NL(T0=T0, t=t, Ts=Ts, k=k.unsqueeze(0))
        generated_data.append(x_hat)
    
    return torch.stack(generated_data)

# Example usage
Ts = torch.tensor(25.0)  # Initial temperature or other relevant start value
t = ti # Time or other domain-specific input
T0 = torch.tensor(15.0)  # Mean of some distribution or parameter, specific to your model setup
new_data = generate_new_data(NL = NewtonsLaw(), num_samples=100, mu_post=mu_tensor.mean(), 
                             var_post=var_tensor.mean(), T0=T0, t=t, Ts=Ts)
new_data = new_data.cpu().detach()
# Convert tensor to numpy
new_data_np = new_data.numpy()

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the size as needed

# Plot each row
for i in range(new_data_np.shape[0]):
    plt.plot(new_data_np[i, :].squeeze(),label=f'Row {i}',  lw=0.4)  # You might want to omit the label if there are many rows

plt.title('Synthetic Data')
plt.xlabel('time')
plt.ylabel('')

# If you have a manageable number of rows, you might include a legend
# plt.legend()

plt.show()
```

```python
def plot_original_vs_reconstructed_grid(original_data, reconstructed_data, num_to_plot=6, grid_cols=3):
    """
    Plot original and reconstructed time series in a grid.

    Parameters:
    - original_data: np.ndarray, original data with shape (num_series, time_steps)
    - reconstructed_data: np.ndarray, reconstructed data with shape (num_series, time_steps)
    - num_to_plot: int, number of time series to plot (default: 6)
    - grid_cols: int, number of columns in the grid (default: 3)
    """
    num_series = min(num_to_plot, original_data.shape[0])  # Limit to available series
    grid_rows = (num_series + grid_cols - 1) // grid_cols  # Calculate number of rows
    time_steps = original_data.shape[1]

    fig, axes = plt.subplots(grid_rows, grid_cols, figsize=(15, grid_rows * 4))
    axes = axes.flatten()  # Flatten the axes for easy iteration

    for i in range(num_series):
        ax = axes[i]
        ax.plot(range(time_steps), original_data[i], label="Original", linestyle='--')
        ax.plot(range(time_steps), reconstructed_data[i], label="Reconstructed", linestyle='-')
        ax.set_title(f"Time Series {i+1}")
        ax.set_xlabel("Time Step")
        ax.set_ylabel("Value")
        ax.legend()
        ax.grid(True)

    # Hide any unused subplots
    for j in range(num_series, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

# Example usage
plot_original_vs_reconstructed_grid(original_data, reconstructed_data, num_to_plot=31, grid_cols=6)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Assuming mu_tensor and var_tensor are given as part of your context

# Define the mean and variance of the underlying normal distribution
mean = mu_tensor.mean().detach().numpy()
variance = var_tensor.mean().detach().numpy()
std_dev = np.sqrt(variance)  # Standard deviation

# Generate a range of x values for the lognormal distribution
x = np.linspace(np.exp(mean - 4*std_dev), np.exp(mean + 4*std_dev), 1000)

# Calculate the corresponding y values for the lognormal distribution
y = lognorm.pdf(x, s=std_dev, scale=np.exp(mean))

# Plot the lognormal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('k')
plt.ylabel('Probability Density')
plt.title('Posterior Distribution')
plt.grid(True)
#plt.legend()
plt.xlim(0, 10)
plt.show()
```
