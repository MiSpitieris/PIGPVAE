```python
# Import necessary libraries
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader


import tsgm
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
file_path = '../Data/RICO4_Dataset_processed.hdf'
df = pd.read_hdf(file_path)
```

```python
df_c = df[df['temp_change_class'] == 'c']
def select_points(group):
    return group.iloc[::10]  # Select every 10th row
df_c = df_c.groupby('Scheduler Step').apply(select_points).reset_index(drop=True)
```

```python
df_c = df_c[['B.RTD1','time_within_interval','interval']]
```

```python
# Initialize a list to collect each time series sequence
sequences = []

# Set the desired sequence length
seq_len = 24

# Group by each unique interval
for _, group in df_c.groupby('interval'):
    # Select B.RTD1 values as the time series data
    sequence = group['B.RTD1'].values
    
    # Truncate or pad the sequence to ensure it has exactly 24 time steps
    if len(sequence) >= seq_len:
        sequence = sequence[:seq_len]
    else:
        sequence = np.pad(sequence, (0, seq_len - len(sequence)), 'constant', constant_values=0)
    
    # Append the sequence as a new entry in the list
    sequences.append(sequence)

# Convert the list of sequences to a DataFrame
df_sequences = pd.DataFrame(sequences)
sequences_array = df_sequences.to_numpy()
sequences_array = sequences_array.reshape((sequences_array.shape[0], sequences_array.shape[1], 1))

# Filter the sequences based on their starting values
train_data = sequences_array[np.array([seq[0] > 24 for seq in sequences_array[:, :, 0]])]
val_data = sequences_array[np.array([seq[0] <= 24 for seq in sequences_array[:, :, 0]])]

print("Training data shape:", train_data.shape)
print("Validation data shape:", val_data.shape)
```

```python
scaler = tsgm.utils.TSFeatureWiseScaler()        
scaled_train_data = scaler.fit_transform(train_data)

# Plot for the first feature
num_sequences = scaled_train_data.shape[0]

plt.figure(figsize=(12, 8))
for i in range(num_sequences):
    plt.plot(scaled_train_data[i, :, 0], label=f'Sequence {i+1}')

plt.title("Time Series Data")
plt.xlabel("Time Steps")
plt.ylabel("B.RTD1 Value")
#plt.legend(loc="upper right", bbox_to_anchor=(1.15, 1))
plt.show()

# original_data = scaler.inverse_transform(scaled_data)
```

```python
model = tsgm.models.timeGAN.TimeGAN(
    seq_len=24,
    module="gru",
    hidden_dim=24,
    n_features=1,  
    n_layers=6,
    batch_size=8,
    gamma=1.0
)


model.compile()
```

```python
epochs=1000
# checkpoints_interval = max(1, epochs // 5)
# generate_synthetic = [int(epochs * 0.2), int(epochs * 0.4), int(epochs * 0.6), int(epochs * 0.8), epochs]

model.fit(
    data=scaled_train_data,
    epochs=epochs
)
```

```python
original_data = scaler.inverse_transform(scaled_train_data)
```

```python
ns = train_data.shape[0]
new_x = model.generate(n_samples=ns)

num_sequences = new_x.shape[0]  # Number of sequences to plot
synthetic = scaler.inverse_transform(new_x)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot original data
for i in range(ns):
    axes[0].plot(scaled_train_data[i, :, 0])
axes[0].set_title("Original Data ")
axes[0].set_xlabel("Time Steps")
axes[0].set_ylabel("Normalized Value")
axes[0].set_ylim(12, 30)

# Plot synthetic data
for i in range(ns):
    axes[1].plot(new_x[i, :, 0])
axes[1].set_title("Generated Synthetic Data")
axes[1].set_xlabel("Time Steps")
axes[1].set_ylabel("Feature Value")
axes[1].set_ylim(12, 30)  # Set y-axis limits

plt.tight_layout()
plt.show()
```
